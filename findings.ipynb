# Imports

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import cross_val_score
```

# Data Cleaning and feature selection

```python
data = pd.read_csv("dataset_37_diabetes.csv")
for i in range(len(data)):
    if data.loc[i,'class']=="tested_positive":
        data.loc[i,'class'] = 1
    else:
        data.loc[i,'class'] = 0
y = data['class']
X = data.drop(columns=['class'])
#X = data.drop(columns=['class', 'pres','skin','insu','pedi'])
print(X.columns)
X
```

# Gradient Boosted Decision Trees Classifier

```python
clf = GradientBoostingClassifier(n_estimators=1000,max_depth=3,learning_rate=0.05)
# Best parameters for maximizing Recall
print("Accuracy --> "+str(np.mean(cross_val_score(clf,X,y,cv=5))))
print("Recall --> "+str(np.mean(cross_val_score(clf,X,y,cv=5,scoring='recall'))))
```

# Random Forest Classifier

```python
clf = RandomForestClassifier(n_estimators=1000,random_state=0) # Best performance at current parameters
print("Accuracy --> "+str(np.mean(cross_val_score(clf,X,y,cv=5))))
print("Recall --> "+str(np.mean(cross_val_score(clf,X,y,cv=5,scoring='recall'))))
```

# Linear Support Vector Classifier

```python
clf = SVC(kernel='linear')
print("Accuracy --> "+str(np.mean(cross_val_score(clf,X,y,cv=5))))
print("Recall --> "+str(np.mean(cross_val_score(clf,X,y,cv=5,scoring='recall'))))
```

# Radial Basis Support Vector Classifier

```python
clf = SVC(kernel='rbf')
print("Accuracy --> "+str(np.mean(cross_val_score(clf,X,y,cv=5))))
print("Recall --> "+str(np.mean(cross_val_score(clf,X,y,cv=5,scoring='recall'))))
```

# K nearest neighbors Classifier

```python
max_i = -1
score = -1
max_i_ = -1
score_ = -1
for i in range(1,20):
    clf = KNeighborsClassifier(n_neighbors=i)
    temp = np.mean(cross_val_score(clf,X,y,cv=5))
    temp2 = np.mean(cross_val_score(clf,X,y,cv=5,scoring='recall'))
    if temp2>score_:
        score_=temp2
        max_i_=i
    if temp>score:
        score = temp
        max_i = i
    print("At k="+str(i)+" accuracy-->"+str(temp))
    print("At k="+str(i)+" recall-->"+str(temp2))
print("***************************************************")
print("Maximum accuracy-->"+str(score)+" at k="+str(max_i))
print("Maximum recall-->"+str(score_)+" at k="+str(max_i_))
```
